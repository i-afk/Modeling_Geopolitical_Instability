```python
import pandas as pd
import glob
import os
import math
import numpy as np
```

```python
################# Importing Files #######################################

path = 'C:\\Users\\Ian\\Desktop\\Ship text' # Specific to me
ship_text = glob.glob(os.path.join(path, "*.txt"))
ship_text

inputs = []
i=0
for file in ship_text: # reads through files in folder
    df = pd.read_csv(file, header = None, sep =" ")
    i=i+1
    df[0] = i 
    inputs.append(df)
    
# Creates Dataframe 
df1 = pd.concat(inputs, sort = False)
df1.set_axis(["image","classes", "x_center", "y_center", "width", "height"], axis='columns', inplace=True)

df1["ship"] = df1.index

df1
```

```python
##################### Final Algorithm ##############

# Grouping 
df2 = df1.groupby("image")


# Proximity 
prox = [] 
for a in range(1, 7, 1): # set for only 7 images 
    x_dist = []
    y_dist = []
    temp = df2.get_group(a) # selects each picture
    w_min = temp["width"].min()   # Creating threshold
    h_min = temp["height"].min()
    if h_min > w_min: # Determines which side of box is width vs height 
        thresh = w_min/2
    else:
        thresh = h_min/2
    for b in range(len(temp)):
        # x
        x = temp.loc[b,"x_center"]
        x_dist.append(temp["x_center"] - x)
        x_squ = np.square(x_dist)
        # y
        y = temp.loc[b,"y_center"]
        y_dist.append(temp["y_center"] - y)
        y_squ = np.square(y_dist)
        # pythag
        mat = x_squ + y_squ
        dist = np.sqrt(mat)
        min_dist =  np.min(dist[np.nonzero(dist)]) # smallest non zero distance 
        if min_dist < thresh:
            prox.append(a)

prox
```

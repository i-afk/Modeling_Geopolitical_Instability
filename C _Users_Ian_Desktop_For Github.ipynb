```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
dgdpr = pd.read_csv("C:\\Users\\Ian\\Desktop\\Macro Debt Cycle\\Debt.GDP.csv")
gdp = pd.read_csv("C:\\Users\\Ian\\Desktop\\Macro Debt Cycle\\GDP.csv")
rgdp = pd.read_csv("C:\\Users\\Ian\\Desktop\\Macro Debt Cycle\\RGDP.csv")
debt = pd.read_csv("C:\\Users\\Ian\\Desktop\\Macro Debt Cycle\\Total_Debt.csv")
```

```python
dgdpr.plot() # interesting, b
```

```python
debt.index= debt["DATE"]
rgdp.index = rgdp["DATE"]
gdp.index = gdp["DATE"]
dgdpr.index = dgdpr["DATE"]
#drgdpr = debt/rgdp
#drgdpr.plot()
```

```python
debt = debt.loc["1/1/1952":"1/1/2022",:]
rgdp = rgdp.loc["1/1/1952":"1/1/2022",:]
gdp = gdp.loc["1/1/1952":"1/1/2022",:]
#dgdpr = dgdpr.loc["1/1/1947":"1/1/2022",:]
type(rgdp)
```

```python
debt = debt["TCMDO"]
rgdp = rgdp["GDPC1"]
gdp = gdp["GDP"]
```

```python
debt = debt.astype(float)
```

```python
data = pd.DataFrame({"Real GDP": rgdp, "GDP": gdp, "Total Debt": debt})
data
```

```python
data["Real GDP"].plot()
```

```python
type(data["GDP"])
```

```python
data["Debt:RDGP Ratio"] = data["Total Debt"] / data["Real GDP"]
    
data["Debt:GDP Ratio"] =  data["Total Debt"] / data["GDP"]
```

```python
data["Debt:RDGP Ratio"].plot()
data["Debt:GDP Ratio"].plot()
plt.legend(["Debt:RDGP Ratio", "Debt:GDP Ratio"], loc= "upper left")
# Very interesting zoom in on the last 20 years
```

```python
dataY2k = data.loc["1/1/2000":"1/1/2022",:]
dataY2k.plot()
```

```python
dataY2k["Debt:RDGP Ratio"].plot()
dataY2k["Debt:GDP Ratio"].plot()
plt.legend(["Debt:RDGP Ratio", "Debt:GDP Ratio"], loc= "upper left")
```
